2x − 3y + z = −1
x − y + 2z = −3

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for:

• Now: solve the above equation using np.linalg.inv function.{Explore more about

Code/Implementation:

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

A_inv = np.linalg.inv(A)
I = np.eye(2)
result = np.dot(A, A_inv)
is_identity = np.allclose(result, I)
print("Is it identity ?  &quot;,is_identity")

AB = np.dot(A, B)
BA = np.dot(B, A)

print("AB:")
print(AB)

print("\nBA:&quot")
print(BA)

is_equal = np.allclose(AB, BA)
print("\nIs AB equal to BA?&quot;, is_equal")

AB_T = np.transpose(np.dot(A, B))
BT_AT = np.dot(np.transpose(B), np.transpose(A))

print("Transpose of AB:")
print(AB_T)

print("\nBT * AT:")
print(BT_AT)
is_equal = np.allclose(AB_T, BT_AT)
print("\nIs (AB)^T equal to B^T * A^T?", is_equal)

A_matrix = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_matrix = np.array([-1, -3, 9])

X = np.linalg.inv(A_matrix).dot(B_matrix)
x, y, z = X

print("Value of x is: {x}")
print("Value of y is: {y}")
print("Value of z is: {z}")

import numpy as np

A_matrix = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_matrix = np.array([-1, -3, 9])

X = np.linalg.solve(A_matrix, B_matrix)

x, y, z = X

print("Value of x is: {x}")
print("Value of y is: {y}")
print("Value of z is: {z}")
```
