```python
# 1. Data Exploration and Understanding:
# - Dataset Overview:
import pandas as pd
#Load the dataset and display the first 10 rows.
dataSet = pd.read_csv('/content/drive/MyDrive/WHR-2024-5CS037.csv')
dataSet.head(10)
```

```python
#• Adding New Columns:
#1. Create a new column called Happiness Category that categorizes countries into three categories based on their Score:
def categorize_data(score):
  if score < 4:
    return 'Low'
  elif 4 <= score <= 6:
    return 'Medium'
  elif score > 6:
    return 'High'

dataSet['Happiness Category'] = dataSet['score'].apply(categorize_data)
print(dataSet)
```

```python
#Task - 3 - Outlier Detection:
import seaborn as sns
Q1_score = southasia_data['score'].quantile(0.25)
Q3_score = southasia_data['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

Q1_gdp = southasia_data['Log GDP per capita'].quantile(0.25)
Q3_gdp = southasia_data['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

outliers = southasia_data[
    (southasia_data['score'] < lower_bound_score) | (southasia_data['score'] > upper_bound_score) |
    (southasia_data['Log GDP per capita'] < lower_bound_gdp) | (southasia_data['Log GDP per capita'] > upper_bound_gdp)
]

print("Outlier Countries:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])

plt.figure(figsize=(8,6))
sns.scatterplot(x='Log GDP per capita', y='score', data=southasia_data, color='blue')
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers, label='Outliers', color='red')
plt.title('Scatter Plot of Happiness Score vs Log GDP per Capita')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

In the scatter plot, outliers are highlighted in red, representing instances with notably lower GDP per Capita and Score. These outliers are typically economically and socially underperforming countries that skew regional averages, potentially giving a misleading impression of South Asia's overall progress. In contrast, the upper range of GDP per Capita and Score reflects stronger performance, largely excluding these outliers.

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

southasia_outliers_score = detect_outliers(southasia_data, 'score')
southasia_outliers_gdp = detect_outliers(southasia_data, 'Log GDP per capita')

middle_east_outliers_score = detect_outliers(middleEast_data, 'score')
middle_east_outliers_gdp = detect_outliers(middleEast_data, 'Log GDP per capita')

plt.subplot(1, 2, 1)
plt.scatter(southasia_data['Log GDP per capita'], southasia_data['score'], color='blue', label='Non-Outliers')
plt.scatter(southasia_outliers_score['Log GDP per capita'], southasia_outliers_score['score'], color='red', label='Outliers (Score)')
plt.scatter(southasia_outliers_gdp['Log GDP per capita'], southasia_outliers_gdp['score'], color='green', label='Outliers (GDP per Capita)')
plt.title("South Asia: Outliers in GDP per Capita and Score")
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.grid()
```
